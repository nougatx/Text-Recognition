```python
# Importing library
import cv2
from pyzbar.pyzbar import decode

# Make one method to decode the barcode 
def BarcodeReader(image):
	
	# read the image in numpy array using cv2
	img = cv2.imread(image)
	
	# Decode the barcode image
	detectedBarcodes = decode(img)
	
	# If not detected then print the message
	if not detectedBarcodes:
		print("Barcode Not Detected or your barcode is blank/corrupted!")
	else:
	
		# Traverse through all the detected barcodes in image
		for barcode in detectedBarcodes: 
		
			# Locate the barcode position in image
			(x, y, w, h) = barcode.rect
			
			# Put the rectangle in image using 
			# cv2 to highlight the barcode
			cv2.rectangle(img, (x-10, y-10),
						(x + w+10, y + h+10), 
						(255, 0, 0), 2)
			
			if barcode.data!="":
			
			# Print the barcode data
				print(barcode.data)
				print(barcode.type)
				
	#Display the image
	cv2.imshow("Image", img)
	cv2.waitKey(0)
	cv2.destroyAllWindows()

if __name__ == "__main__":
# Take the image from user
	image="ss.png"
	BarcodeReader(image)
```

```python
import cv2

camera_id = 1
delay = 1
window_name = 'OpenCV Barcode'

bd = cv2.barcode.BarcodeDetector()
cap = cv2.VideoCapture(camera_id)

while True:
    ret, frame = cap.read()

    if ret:
        ret_bc, decoded_info, decoded_points = bd.detectAndDecode(frame)
        if ret_bc:
            if decoded_points is not None:
                frame = cv2.polylines(frame, decoded_points.astype(int), True, (0, 255, 0), 3)
            if isinstance(decoded_info, list):
                for s, p in zip(decoded_info, decoded_points):
                    if s:
                        print(s)
                        frame = cv2.putText(frame, s, p[1].astype(int),
                                            cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 2, cv2.LINE_AA)
            else:
                print(decoded_info)
                frame = cv2.putText(frame, decoded_info, (10, 50),
                                    cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 2, cv2.LINE_AA)

        cv2.imshow(window_name, frame)

    if cv2.waitKey(delay) & 0xFF == ord('q'):
        break

cv2.destroyWindow(window_name)
```

```python
import cv2
from pyzbar.pyzbar import decode
import numpy as np

def decode_barcode(image_path):
    # Read the image
    image = cv2.imread(image_path)

    # Convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Decode the barcode
    barcodes = decode(gray)

    if barcodes:
        for barcode in barcodes:
            # Extract the barcode data
            barcode_data = barcode.data.decode("utf-8")

            # Print the decoded barcode information
            print("Decoded Barcode:")
            print(f"Data: {barcode_data}")
            print(f"Type: {barcode.type}")
            print(f"Bounding Box: {barcode.rect}")
            print("")

            # Draw a rectangle around the barcode
            points = barcode.polygon
            if len(points) == 4:
                pts = []
                for p in points:
                    pts.append([p.x, p.y])
                pts = np.array(pts, dtype=int)
                image = cv2.polylines(image, [pts], True, (0, 255, 0), 2)

        # Display the image with the barcode
        cv2.imshow("Decoded Barcode", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

    else:
        print("No barcode detected in the image.")

# Example usage
image_path = "sophie.jpeg"
decode_barcode(image_path)
```
